```python
# %pip install --upgrade pip
# %pip install opencv-python
import cv2
import os.path
import tensorflow as tf
from ipywidgets import IntProgress
import IPython.display
import random
import matplotlib.pyplot as plt
import seaborn as sns
input_size = (512,512,3)
classes_number = 4

SAVE_PATH="./models/version4"
TF_RECORD_DATA=[
    "./train_1.tfrecord"
]
```

```python
model=None
trained=False
if os.path.exists(SAVE_PATH):
    print("load model from files")
    model = tf.keras.models.load_model(SAVE_PATH)
else:
    print("build new model")
    #Model Defining
    model=tf.keras.Sequential()

    model.add(tf.keras.layers.Conv2D(32,kernel_size=(3,3),
                    activation='relu',
                    input_shape=input_size,
                    kernel_initializer='he_normal'))  
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    model.add(tf.keras.layers.Conv2D(32,kernel_size=(3,3),
                    activation='relu',
                    kernel_initializer='he_normal'))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(3,3)))
    model.add(tf.keras.layers.Conv2D(64,kernel_size=(3,3),
                    activation='relu',
                    kernel_initializer='he_normal'))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    model.add(tf.keras.layers.Conv2D(128,kernel_size=(3,3),
                    activation='relu',
                    kernel_initializer='he_normal'))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    model.add(tf.keras.layers.Conv2D(64,kernel_size=(3,3),
                    activation='relu',
                    kernel_initializer='he_normal'))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    model.add(tf.keras.layers.Conv2D(32,kernel_size=(3,3),
                    activation='relu',
                    kernel_initializer='he_normal'))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(classes_number,activation='softmax',kernel_initializer='glorot_normal'))
    #Model Compile
    model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
    
model.summary()
```

```python
# Create a dictionary describing the features.
image_feature_description = {
    'height': tf.io.FixedLenFeature([], tf.int64),
    'width': tf.io.FixedLenFeature([], tf.int64),
    'depth': tf.io.FixedLenFeature([], tf.int64),
    'label': tf.io.FixedLenFeature([], tf.int64),
    'image_raw': tf.io.FixedLenFeature([], tf.string),
}

def _show_image_function(example):
    example = tf.io.parse_single_example(example, image_feature_description)
    return tf.image.decode_jpeg(example["image_raw"])

def _parse_image_function(example_proto):
    # Parse the input tf.train.Example proto using the dictionary above.
    example = tf.io.parse_single_example(example_proto, image_feature_description)
    return [ tf.image.per_image_standardization(tf.image.resize(tf.image.decode_jpeg(example["image_raw"]), [input_size[0], input_size[1]])), tf.one_hot(example["label"], classes_number)]
```

```python
for img in tf.data.TFRecordDataset(TF_RECORD_DATA).map(_show_image_function):
    plt.imshow(img)
    break
```

```python
full_dataset = tf.data.TFRecordDataset(TF_RECORD_DATA)
full_dataset = full_dataset.map(_parse_image_function)
```

```python
full_dataset = full_dataset.shuffle(3000)
train_dataset = full_dataset.batch(80)
```

```python
trained=False
epoch=100
if not trained:
    his=model.fit(x=train_dataset, epochs=epoch, verbose=1)
    
else:
    print("dataset trained, skip this step")
```

```python
f = plt.figure(figsize=(20,7))

#Adding Subplot 1 (For Accuracy)
f.add_subplot(121)
Title_plt_acc='_accuracy'
Title_plt_loss='training_Loss'
sns.set_style("darkgrid")
plt.plot(his.epoch,his.history['accuracy'],label = "train_accuracy") # Accuracy curve for training set
# plt.plot(his.epoch,his.history['val_accuracy'],label = "val_accuracy") # Accuracy curve for validation set

plt.title(Title_plt_acc,fontsize=18)
plt.xlabel("Epochs",fontsize=15)
plt.ylabel("Accuracy",fontsize=15)
plt.grid(alpha=0.3)
# plt.yticks([0.960,0.965,0.970,0.975,0.980,0.985,0.990,0.995,1])
plt.legend()

f.add_subplot(122)
sns.set_style("darkgrid")
plt.plot(his.epoch,his.history['loss'],label="train_loss") # Loss curve for training set
# plt.plot(his.epoch,his.history['val_loss'],label="val_loss") # Loss curve for validation set

plt.title(Title_plt_loss,fontsize=18)
plt.xlabel("Epochs",fontsize=15)
plt.ylabel("Loss",fontsize=15)
plt.grid(alpha=0.3)
plt.yticks([0,0.0005,0.001,0.0015,0.002])
plt.savefig("loss1.png")
plt.legend()

# plt.savefig(Save_file_dataside+"chart.png")
plt.show()
```
